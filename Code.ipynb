```python
def find_memory(x):
  res = re.findall(r'(?<=memory )\d+ [g|m]', x.lower())
  if len(res) == 0:
    return 0
  else:
    m = 0
    for i in res:
      if i[-1] == 'g':
        m += int(i[:-2]) * 1024
      elif i[-1] == 'm':
        m += int(i[:-2])
    return m/len(res)
    
steam_requirements_data['memory'] = steam_requirements_data['minimum'].apply(lambda x: find_memory(x))
```

```python
def find_storage(x):
  res = re.findall(r'(?<=storage )\d+ [g|m]', x.lower())
  if len(res) == 0:
    return 0
  else:
    m = 0
    for i in res:
      if i[-1] == 'g':
        m += int(i[:-2]) * 1024
      elif i[-1] == 'm':
        m += int(i[:-2])
    return m/len(res)
    
steam_requirements_data['storage'] = steam_requirements_data['minimum'].apply(lambda x: find_storage(x))
```
